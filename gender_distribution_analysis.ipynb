```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# If reading from file, use: df = pd.read_excel("UMERA BOOST 2.0 & 3.0.xlsx", sheet_name="UMRA 3.0")
df = pd.read_excel("UMERA BOOST 2.0 & 3.0.xlsx", sheet_name="UMRA 3.0")

# Clean column names to remove hidden characters and whitespace
df.columns = df.columns.str.strip().str.replace(r'\u2060', '', regex=True)

# Define the gender column
gender_column = 'Gender'

# Handling missing values in Gender
df[gender_column] = df[gender_column].fillna("Unknown")

# Count the number of participants by gender
gender_distribution = df[gender_column].value_counts()

# Visualization
plt.figure(figsize=(8, 6))
# sns.barplot(x=gender_distribution.index, y=gender_distribution.values, palette='coolwarm')
sns.barplot(x=gender_distribution.index, y=gender_distribution.values, hue=gender_distribution.index, palette='coolwarm', legend=False)
plt.xlabel("Gender")
plt.ylabel("Number of Participants")
plt.title("Gender Distribution of Registrants")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

#Pie chart visualization
plt.figure(figsize=(8, 8))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', 
        startangle=90, colors=['blue', 'pink'], wedgeprops={'edgecolor': 'black'})
plt.title("Gender Distribution of Registrants")
plt.show()


# Display numerical distribution
print("Gender Distribution:\n")
print(gender_distribution)

# Percentage breakdown
gender_percentage = (gender_distribution / gender_distribution.sum()) * 100
print("\nPercentage Breakdown of Gender:\n")
print(gender_percentage.round(2).astype(str) + '%')
```
